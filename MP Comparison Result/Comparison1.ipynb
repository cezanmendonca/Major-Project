```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the evaluation metrics data
data = {
    "Algorithm": ["SGD", "Hybrid", "AdaBoost"],
    "Training Accuracy": [0.761152, 0.998754, 0.999635],
    "Test Accuracy": [0.745967, 0.998725, 0.999675],
    "Optimal Test Accuracy": [0.754993, 0.998918, 0.999675],
    "Sensitivity": [0.76206, 0.998136, 0.999596],
    "Specificity": [0.75787, 0.999314, 0.999754],
    "Precision": [0.758883, 0.999313, 0.999754],
    "F1 Score": [0.760468, 0.998724, 0.999675],
    "MCC": [0.519938, 0.997451, 0.999349],
    "BCR": [0.759967, 0.998725, 0.999675],
    "ROC AUC": [0.76006, 0.999965, 0.999968],
    "Average Precision": [0.692371, 0.999973, 0.999926],
    "Cohen’s Kappa": [0.519434, 0.99745, 0.999349],
    "Log Loss": [8.651668, 0.142256, 0.300422],
    "Training Time (s)": [234.30, 149.78, 136.71],
    "Optimal Threshold": [8.086462e-303, 0.449075, 0.502309]
}
```

```python
# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Display the table
print("Comparison Table of Evaluation Metrics:")
display(df)

# Plot comparison graphs for each metric
metrics = df.columns[1:-1]  # Exclude "Algorithm" and "Optimal Threshold"

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20, 20))
axes = axes.flatten()

for i, metric in enumerate(metrics):
    ax = axes[i]
    df.plot(x="Algorithm", y=metric, kind="bar", ax=ax, legend=False, colormap="viridis")
    ax.set_title(metric)
    ax.set_ylabel(metric)

# Hide unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
